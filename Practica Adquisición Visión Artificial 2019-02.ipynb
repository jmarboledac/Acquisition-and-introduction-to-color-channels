Separe y muestre las imágenes en los canales R,G,B del espacio de color RGB. Realice el mismo proceso para el espacio de color LAB, mostrando los canales L,A y B* respectivamente. ¿En qué canal se observa más fácilmente el objeto de estudio? ¿Depende esto de la fotografía que se estudie?
```python
%matplotlib inline

import cv2
import matplotlib.pyplot as plt
import numpy as np
#Leer y plotear la imagen de trabajo
i7 = cv2.imread("res/canales.jpg", cv2.IMREAD_COLOR)
img_rgb  = cv2.cvtColor(i7, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_lab  = cv2.cvtColor(i7, cv2.COLOR_BGR2LAB)
plt.imshow(img_lab)
```

```python
img_R = img_rgb[:,:,0]
img_G = img_rgb[:,:,1]
img_B = img_rgb[:,:,2]
img_L = img_lab[:,:,0]
img_A =  img_lab[:,:,1]
img_B = img_lab[:,:,2]
#Canal RGB vs LAB
fig, ((a1, a2, a3), (a4, a5, a6)) = plt.subplots(2,3, figsize=(28, 15))
fig.suptitle('RBG vs LAB', fontsize=20)
a1.set_title('Canal R del RGB')
a1.imshow(img_R, cmap='Reds', aspect='auto')
a2.set_title('Canal G del RGB')
a2.imshow(img_G, cmap='Greens', aspect='auto')
a3.set_title('Canal B del RGB')
a3.imshow(img_B, cmap='Blues', aspect='auto')
a4.set_title('Canal Luminosidad del LAB')
a4.imshow(img_L, cmap='gray', aspect='auto')
a5.set_title('Canal A* del LAB')
a5.imshow(img_A, cmap='gray', aspect='auto')
a6.set_title('Canal B* del LAB')
a6.imshow(img_B, cmap='gray', aspect='auto')
#se eligió el canal b del LAB porque se usa para diferenciar de azul a amarillo. El azul se va a ver negro y el amarillo blanco
#por lo que se logra diferenciar facilmente las pequeñas partes amariilas#
```

## Actividad 3.

Una vez escogida la imagen donde mejor se puede apreciar el objeto, en el canal del espacio de color respectivo, ¿Puede separar usted el objeto del fondo?
<br> Genere una segmentación de tipo binario y analice. </br>

```python
ret,thresh2 = cv2.threshold(img_B,150,255,cv2.THRESH_BINARY)
plt.imshow(thresh2, cmap='gray')
```

```python
from skimage import morphology as morph
thresh2 = thresh2/255
res = morph.remove_small_holes(thresh2.astype(bool), min_size = 100000)
res = morph.remove_small_objects(res, min_size = 2000)
plt.imshow(res)
```

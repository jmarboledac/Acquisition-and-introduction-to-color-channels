```python
# Librerías
import cv2
import matplotlib.pyplot as plt
import numpy as np
#Lectura de la imagen
fruta = cv2.imread("res/frutas.jpg", cv2.IMREAD_COLOR)
#Descomposición en modelos
f_rgb  = cv2.cvtColor(fruta, cv2.COLOR_BGR2RGB)
f_lab  = cv2.cvtColor(fruta, cv2.COLOR_BGR2LAB)
f_hsv  = cv2.cvtColor(fruta, cv2.COLOR_BGR2HSV)
f_yuv  = cv2.cvtColor(fruta, cv2.COLOR_BGR2YUV)
f_xyz  = cv2.cvtColor(fruta, cv2.COLOR_BGR2XYZ)

#Columnas y filas
fig, arreglo_plots = plt.subplots(5,3, figsize=(15, 15))
colors  = ('r','g','b')
colors1 = ('r','g','b')
colors2 = ('l','a','b')
colors3 = ('h','v','s')
colors4 = ('y','u','v')
colors5 = ('x','y','z')

#Imagen en RGB
I1_array=f_rgb.ravel()
arreglo_plots[0,0].set_title('Imagen en RGB')
arreglo_plots[0,0].imshow(f_rgb, cmap="gray")
arreglo_plots[0,1].set_title('Histograma general')
_ = arreglo_plots[0,1].hist(I1_array, bins=255, range=(0.0, 255.0),log=True)
arreglo_plots[0,2].set_title('Histograma por canales')
for i,col in enumerate(colors1):
    img_array_i = f_rgb[:,:,i].ravel()
    arreglo_plots[0,2].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])
    
#Imagen en LAB
I2_array=f_lab.ravel()
arreglo_plots[1,0].set_title('Imagen en LAB')
arreglo_plots[1,0].imshow(f_lab, cmap="gray")
_ = arreglo_plots[1,1].hist(I2_array, bins=255, range=(0.0, 255.0),log=True)
for i,col in enumerate(colors2):
    img_array_i = f_lab[:,:,i].ravel()
    arreglo_plots[1,2].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])
    
#Imagen en HSV
I3_array=f_hsv.ravel()
arreglo_plots[2,0].set_title('Imagen en HSV')
arreglo_plots[2,0].imshow(f_hsv, cmap="gray")
_ = arreglo_plots[2,1].hist(I3_array, bins=255, range=(0.0, 255.0),log=True)
for i,col in enumerate(colors3):
    img_array_i = f_hsv[:,:,i].ravel()
    arreglo_plots[2,2].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])
    
#Imagen en YUV
I4_array=f_yuv.ravel()
arreglo_plots[3,0].set_title('Imagen en YUV')
arreglo_plots[3,0].imshow(f_yuv, cmap="gray")
_ = arreglo_plots[3,1].hist(I4_array, bins=255, range=(0.0, 255.0),log=True)
for i,col in enumerate(colors4):
    img_array_i = f_yuv[:,:,i].ravel()
    arreglo_plots[3,2].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])
    
#Imagen en XYZ
I5_array=f_xyz.ravel()
arreglo_plots[4,0].set_title('Imagen en XYZ')
arreglo_plots[4,0].imshow(f_xyz, cmap="gray")
_ = arreglo_plots[4,1].hist(I5_array, bins=255, range=(0.0, 255.0),log=True)
for i,col in enumerate(colors5):
    img_array_i = f_xyz[:,:,i].ravel()
    arreglo_plots[4,2].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])
```

```python
# Actividad 4

%matplotlib inline

import cv2
from matplotlib import pyplot as plt
import numpy as np

img1 =cv2.cvtColor(cv2.imread('res/gato.jpg'), cv2.COLOR_BGR2RGB)

edges = cv2.Canny(img1,50,100)

edges[edges<128] = 0   

edges = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)

edges[:,:,2]=0            


fig = plt.figure(figsize=(15,24))

plt.subplot(1,3,1)
plt.imshow(img1,cmap='gray')
plt.title("Imagen 1")

plt.subplot(1,3,2)
plt.imshow(edges,cmap='gray')
plt.title("Imagen 2")

plt.subplot(1,3,3)
plt.imshow(cv2.add(img1,edges),cmap='gray')
plt.title("Imagen 3");
```

```python
# Actividad 5

import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread('res/feynman.jpg',cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

Ancho=img.shape[1]
Alto=img.shape[0]
Tamaño="Tamaño original: " + str(Ancho)+"x"+str(Alto)

imagen1 = cv2.resize(img,(900,1600), interpolation = cv2.INTER_CUBIC)
imagen2 = cv2.resize(img,(600,800), interpolation = cv2.INTER_CUBIC)
imagen3 = cv2.resize(img,(250,250), interpolation = cv2.INTER_CUBIC)

fig = plt.figure(figsize=(12,8), dpi = 300)

plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title(Tamaño)

plt.subplot(2,2,2)
plt.imshow(imagen1,cmap='gray')
plt.title("1600x900")

plt.subplot(2,2,3)
plt.imshow(imagen2,cmap='gray')
plt.title("800x600")

plt.subplot(2,2,4)
plt.imshow(imagen3,cmap='gray')
plt.title("250x250")

imagen1 = cv2.cvtColor(imagen1, cv2.COLOR_BGR2RGB)
imagen2 = cv2.cvtColor(imagen2, cv2.COLOR_BGR2RGB)
imagen3 = cv2.cvtColor(imagen3, cv2.COLOR_BGR2RGB)

cv2.imwrite("imagen1.jpg", imagen1);
cv2.imwrite("imagen2.jpg", imagen2);
cv2.imwrite("imagen3.jpg", imagen3);
```

```python
# Aclaración punto 5

print("Se utilizo solamente la transformación bicúbica debido a las ventajas que esta tiene sobre las otras. Esta transformación permite la generación de imagenes más nítidas, dando un buen compromiso entre el tiempo de procesado y la calidad del resultado.")
```
